# Model Selection

1. As the problem is of Regression and data is small Linear regression is implemented.

2. As Data is small and can have some Non Linearity SVM is implemented.

3. As data is small KNN is implemented.

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train) # Model Training
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)**0.5
```

```python
r2_score(y_test,y_pred)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```



# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
kn=KNeighborsRegressor(n_neighbors=6)
```

```python
kn.fit(X_train,y_train)
```

```python
y_pred=kn.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)*100
```

# SVM

```python
from sklearn.model_selection import GridSearchCV
```

```python
param={'kernel':['linear','poly','rbf','sigmoid']}
```

```python
from sklearn.svm import SVR
```

```python
sr=SVR()
```

```python
gsc = GridSearchCV(estimator=sr,
                   param_grid = param,
                   scoring = 'r2'
                   )
```

```python
gsc.fit(X_train,y_train)
```

```python
y_pred=gsc.best_estimator_.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
